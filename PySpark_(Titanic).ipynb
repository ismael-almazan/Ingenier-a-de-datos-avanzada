## 1. Carga de datos
Carga el dataset Titanic usando spark.read.csv

```python
df = spark.read.csv('/content/Titanic-Dataset.csv', header=True, inferSchema=True) # Con inferSchema en True para identificar los tipos de detos en automatico
```

## 2. Exploración inicial
Muestra las primeras filas con show()

Visualiza el esquema con printSchema()

```python
#Primeras columas
df.show(7)


#Visualizar el esquemas donde se muestra los nmbre de las columnas y el tipo de dato
df.printSchema()
```

## 3. Selección de columnas
Selecciona las columnas: Name, Age, Sex

Muestra el resultado

```python
# Seleccionar columas Nane, Age y Sex
col = ['Name', 'Age', 'Sex']

# Mostrar el resultado
df.select(col).show()
```

## 4. Filtrado de datos
Filtra los pasajeros:

Mayores de 18 años

De sexo femenino

Muestra los resultados

```python
#Filtrar Pasajeros mayores de 18 anos, femeninos
df.filter("Age > 18 AND Sex = 'female'").show()
```

## 5. Valores únicos
Obtén los valores únicos de la columna Pclass

```python
df.select("Pclass").distinct().show()  #Muestra las tres clases
```

## 6. Agrupación
Cuenta el número de pasajeros por:

Pclass

Sex


```python
df.groupBy("Pclass", "Sex").count().show()  # Agrupa las clases123 de los pasajeros por sexo
```

## 7. Estadística básica
Aplica describe() sobre las variables numéricas


```python
df.select("Age", "Fare", "SibSp", "Parch").describe().show()   #Tabla de estadistica basica de las col numericas, conteo, media, desv.est, min, max
```

## 8. Valores nulos
Identifica en qué columnas hay valores nulos (usa isNull() o funciones vistas)

```python
# Mostra los nombre de las columanas que contiene valores nulos y el numero de valores nulos con un ciclo
for columna in df.columns:
    conteo_nulos = df.filter(df[columna].isNull()).count()
    if conteo_nulos > 0:
        print(f"Columna '{columna}': {conteo_nulos} valores nulos")
```
