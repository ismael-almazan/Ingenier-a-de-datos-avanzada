**Objetivo**
Aplicar transformaciones, acciones y operaciones matemáticas sobre un RDD generado a partir de datos simulados.

**Planteamiento**
Genera una lista de datos numéricos utilizando una distribución normal aleatoria en Python. Posteriormente, convierte estos datos en un RDD y realiza diferentes operaciones para analizarlos.





```python
# Lbrerias requeridas
import numpy as np
import math
import pyspark

from pyspark.sql import SparkSession

print("Versión pyspark:", pyspark.__version__)
```

# 1. Genera una lista de al menos 1000 valores usando una función de Python.

Usa una distribución normal con parámetros controlados para que los resultados sean interpretables:

*  Media (μ): 50

*  Desviación estándar (σ): 10

*  Tamaño de muestra: 1000 valores

*  Fijar semilla a 18


```python
# Parametros para generar la distribucion
mu = 50
sigma = 10
n = 1000
semilla = 18

# Generacion de datos
np.random.seed(semilla)
lista = np.random.normal(mu, sigma, n)

print(lista[0:10])
```

# 2. Paraleliza la lista para crear un RDD llamado rdd0

```python
#Crear SparkSessioN
spark = SparkSession.builder \
    .appName("RDD IDA") \
    .getOrCreate()



# Crear el SparkContext
sc = spark.sparkContext

print(f"\n✓ SparkContext creado exitosamente")
print(f"  - Nombre de la aplicación: {sc.appName}")
print(f"  - Master: {sc.master}")
print(f"  - Versión de Spark: {sc.version}")
print(f"  - Python Version: {sc.pythonVer}")
print(f"  - Default Parallelism: {sc.defaultParallelism}")


#Paralelizar la lista
rdd0=sc.parallelize(lista.tolist())    #tolist para que solo se carguen los valores sin tu tipo de dstos
print(rdd0)
print("Parciciones", rdd0.getNumPartitions())
```

# 3. Aplica las siguientes operaciones:
a) Transformaciones

* Usa map() para elevar al cuadrado cada elemento.

* Usa take(5) para mostrar 5 resultados

* Usa filter() para quedarte solo con valores que sean mayores a 50. Imprime en pantalla el resultado de 4 resultados.

b) Transformaciones en Pseudo-Conjuntos

* Genera dos conjuntos de datos (listas) de 1000 valores cada uno utilizando una distribución normal con:

  * Media (μ) = 50

  * Desviación estándar (σ) = 10

* Convierte ambas listas en RDDs:

  * rdd1

  * rdd2

* Aplica las siguientes transformaciones de pseudo-conjuntos:

  * Usa union() para combinar ambos RDDs.

  * Usa distinct() para eliminar valores duplicados del resultado anterior.

  * Usa intersection() para obtener los valores comunes entre rdd1 y rdd2.

  * Usa subtract() para obtener los valores que están en rdd1 pero no en rdd2.

  * Usa cartesian() para generar todas las posibles combinaciones entre ambos RDDs.

* Ejecuta acciones para observar los resultados:

  * Usa take() o takeOrdered() para visualizar 5 elementos de cada transformación.

  * Usa count() para conocer el tamaño de cada RDD resultante.



```python
#Transformaciones---------------------------------------------------------
print("TRANSFORMACIONES\n")

  #Cuadrado de cada elemento
datos_cuadrado = rdd0.map(lambda x: x**2)
print("Cuadrado de cada elemento\n", datos_cuadrado.take(5))
print("Tamano del rdd:", datos_cuadrado.count())

  #Valores mayores a 50
datos_mayores50 = rdd0.filter(lambda x: x > 50)
print("\nValores mayores a 50\n", datos_mayores50.take(4))
print("Tamano del rdd:", datos_mayores50.count())




#Transformaciones en Pseudo-Conjuntos-------------------------------------
print("\n\n\nTRANSFORMACIONES EN PSEUDO-CONJUNTOS\n")

  #Genarar los conjuntos de datos lista1 y lista2
lista1=np.random.normal(mu, sigma, n)
lista2=np.random.normal(mu, sigma, n)



  #Convertir listas a rdds
rdd1=sc.parallelize(lista1.tolist())
rdd2=sc.parallelize(lista2.tolist())



  #Traformaciones a rdd1 y rrd2
    #Union de rdd1 y rdd2
rdd_union=rdd1.union(rdd2)
print("Union\n", rdd_union.take(5))
print("Tamano del rdd:", rdd_union.count())

    #Distinct
rdd_distinct=rdd_union.distinct()
print("\nDistinct (Valores Unicos)\n", rdd_distinct.take(5))
print("Tamano del rdd:", rdd_distinct.count())

    #Intersection
rdd_intersection=rdd1.intersection(rdd2)
print("\nIntersection\n", rdd_intersection.take(5))
print("Tamano del rdd:", rdd_intersection.count())

    #subtract
rdd_subtract=rdd1.subtract(rdd2)
print("\nsubtract(valores en rdd1 pero no en rdd2)\n", rdd_subtract.take(5))
print("Tamano del rdd:", rdd_subtract.count())

    #Cartesian
rdd_cartesian=rdd1.cartesian(rdd2)
print("\nCartesian(ombinaciones entre ambos RDDs)\n", rdd_cartesian.take(5))
print("Tamano del rdd:", rdd_cartesian.count())
```

# 4. Operaciones matemáticas y estadísticas.

* Utiliza rdd0 y calcula el promedio, valor máximo y mínio.
* Calcula la varianza y desviación estándar
* Utiliza la biblioteca math y calcula la raíz cuadrada. Imprime solo 5 resultados

```python
# Promeido, valor maximo y minimo
print("Valor maximo:", rdd0.max())
print("Valor minimo:", rdd0.min())
print("Promedio:", rdd0.mean())

# Varinza y desv. est
print("\nVarianza:", rdd0.variance())
print("Desviacion estandar:", rdd0.stdev())

print("\nRaiz cuadrada:", math.sqrt(rdd0.mean()))
```
